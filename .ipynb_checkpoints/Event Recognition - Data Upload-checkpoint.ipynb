```python
# Imports
import os
from shutil import copyfile
import pathlib as p

import pandas as pd
```

```python
# Args
event_table_path = p.Path("/dls/science/groups/i04-1/conor_dev/pandda_nn/new_events_no_cheat.csv")
hierarchy_root_path = p.Path("/dls/science/groups/i04-1/conor_dev/data")
```

```python
# Load dataset
event_table = pd.read_csv(str(event_table_path)).set_index("Unnamed: 0")
```

```python
event_table
```

```python
# Set up output directory
try:
    os.mkdir(str(hierarchy_root_path))
except Exception as e:
    print(e)
```

```python
# Iterate over datasets
for record in event_table.iterrows():
    index = record[0]
    record = record[1]    
    
    # Make system path
    system_path = hierarchy_root_path / record.system
    try:
        os.mkdir(str(system_path))
        print("Made directoy for system {}".format(record.system))
    except Exception as e:
        print("{}".format(e))
    
    # Make dataset path
    dataset_path = system_path / record.dtag
    try:
        os.mkdir(str(dataset_path))
        print("\tMade directory for system {}".format(record.dtag))
    except Exception as e:
        print("\t{}".format(e))

    # Copy each file into a sensible hierarchy
    average_map_path = dataset_path / "ground_map_{}.ccp4".format(index)
    try:
        copyfile(str(record.ground_map_path), str(average_map_path))
        print("\t\tCopied ground map for dataset {}, index {}".format(record.dtag, index))
    except Exception as e:
        print("\t\t{}".format(e))
        
    z_map_path = dataset_path / "z_map_{}.ccp4".format(index)
    try:
        copyfile(str(record.z_map_path), str(z_map_path))
        print("\t\tCopied z map for dataset {}, index {}".format(record.dtag, index))

    except Exception as e:
        print("\t\t{}".format(e))
        
    event_map_path = dataset_path / "event_map_{}.ccp4".format(index)
    try:
        copyfile(str(record.event_map_path), str(event_map_path))
        print("\t\tCopied event map for dataset {}, index {}".format(record.dtag, index))
    except Exception as e:
        print("\t\t{}".format(e))
```
